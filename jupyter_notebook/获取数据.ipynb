  ---
  数据采集优先级建议

  根据可行性和对模型的贡献度排序：

  1. 高优先级（立即可获取，零门槛）：AKShare 财报/行情/宏观 — 这是整个模型的基础
  2. 中优先级（需要积分或注册）：Tushare Pro 部分数据、新闻爬虫
  3. 低优先级（需手工整理或爬取）：供应链关系（年报PDF解析）、关税历史节点

  关键提醒：AKShare 是目前国内金融数据采集最友好的免费接口，覆盖了文档中绝大多数需求，且无需申请积分。Tushare Pro
  作为补充（尤其新闻接口更稳定），但需要注册并积累积分。建议优先以 AKShare 为主数据源启动。
